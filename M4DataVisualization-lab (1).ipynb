# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt
# print how many rows are there in the table named 'master'
QUERY = """
SELECT ConvertedComp
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df_cc = pd.read_sql_query(QUERY,conn)
df_cc.shape

plt.figure(figsize=(10, 6))
plt.hist(df_cc['ConvertedComp'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of ConvertedComp')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df_age = pd.read_sql_query(QUERY,conn)
print(df_age)

# Remove NaN values from the 'Age' column
df_age_cleaned = df_age.dropna(subset=['Age'])

# Plot the box plot of the cleaned 'Age' column
plt.figure(figsize=(10, 6))
plt.boxplot(df_age_cleaned["Age"], vert=False, patch_artist=True, notch=True)
plt.title('Box plot of Age (Without NaN Values)')
plt.xlabel('Age')
plt.grid(True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
print(df)

# Calculate the IQR (Interquartile Range) for 'Age' and 'WorkWeekHrs' columns
Q1_age = df['Age'].quantile(0.25)
Q3_age = df['Age'].quantile(0.75)
IQR_age = Q3_age - Q1_age

Q1_work_week_hrs = df['WorkWeekHrs'].quantile(0.25)
Q3_work_week_hrs = df['WorkWeekHrs'].quantile(0.75)
IQR_work_week_hrs = Q3_work_week_hrs - Q1_work_week_hrs

# Define the lower and upper bounds for outliers removal
lower_bound_age = Q1_age - 1.5 * IQR_age
upper_bound_age = Q3_age + 1.5 * IQR_age

lower_bound_work_week_hrs = Q1_work_week_hrs - 1.5 * IQR_work_week_hrs
upper_bound_work_week_hrs = Q3_work_week_hrs + 1.5 * IQR_work_week_hrs

# Remove outliers from the DataFrame
df_cleaned = df[(df['Age'] >= lower_bound_age) & (df['Age'] <= upper_bound_age) &
                (df['WorkWeekHrs'] >= lower_bound_work_week_hrs) & (df['WorkWeekHrs'] <= upper_bound_work_week_hrs)]

# Create a scatter plot of the cleaned data
plt.figure(figsize=(10, 6))
plt.scatter(df_cleaned['Age'], df_cleaned['WorkWeekHrs'], alpha=0.5)
plt.title('Scatter Plot of Age and WorkWeekHrs (Without Outliers)')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)

# Calculate the IQR (Interquartile Range) for 'WorkWeekHrs' and 'CodeRevHrs' columns
Q1_work_week_hrs = df['WorkWeekHrs'].quantile(0.25)
Q3_work_week_hrs = df['WorkWeekHrs'].quantile(0.75)
IQR_work_week_hrs = Q3_work_week_hrs - Q1_work_week_hrs

Q1_code_rev_hrs = df['CodeRevHrs'].quantile(0.25)
Q3_code_rev_hrs = df['CodeRevHrs'].quantile(0.75)
IQR_code_rev_hrs = Q3_code_rev_hrs - Q1_code_rev_hrs

# Define the lower and upper bounds for outliers removal
lower_bound_work_week_hrs = Q1_work_week_hrs - 1.5 * IQR_work_week_hrs
upper_bound_work_week_hrs = Q3_work_week_hrs + 1.5 * IQR_work_week_hrs

lower_bound_code_rev_hrs = Q1_code_rev_hrs - 1.5 * IQR_code_rev_hrs
upper_bound_code_rev_hrs = Q3_code_rev_hrs + 1.5 * IQR_code_rev_hrs

# Remove outliers from the DataFrame
df_cleaned = df[(df['WorkWeekHrs'] >= lower_bound_work_week_hrs) & (df['WorkWeekHrs'] <= upper_bound_work_week_hrs) &
                (df['CodeRevHrs'] >= lower_bound_code_rev_hrs) & (df['CodeRevHrs'] <= upper_bound_code_rev_hrs)]

# Create a bubble plot of the cleaned data
plt.figure(figsize=(10, 6))
plt.scatter(df_cleaned['WorkWeekHrs'], df_cleaned['CodeRevHrs'], s=df_cleaned['Age'], alpha=0.5)
plt.title('Bubble Plot of WorkWeekHrs and CodeRevHrs (Age as Bubble Size)')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.grid(True)
plt.show()
```

### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) AS OccurrenceCount
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY OccurrenceCount DESC
LIMIT 5;
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)


# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(df['OccurrenceCount'], labels=df['DatabaseDesireNextYear'], autopct='%1.1f%%')
plt.title('Top 5 Most Popular Databases')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age, ConvertedComp, MainBranch
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)

# Step 1: Filter the DataFrame for respondents aged 30 to 35
filtered_df = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Step 2: Group by age and calculate the median WorkWeekHrs and CodeRevHrs
grouped_df = filtered_df.groupby('Age').median()[['WorkWeekHrs', 'CodeRevHrs']]

# Step 3: Plot the stacked bar chart
grouped_df.plot(kind='bar', stacked=True)
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.xticks(rotation=0)
plt.legend(title='Metrics')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# Step 1: Filter the DataFrame for respondents aged 45 to 60
filtered_df = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Step 2: Group by age and calculate the median ConvertedComp
grouped_df = filtered_df.groupby('Age').median()['ConvertedComp']

# Step 3: Plot the median ConvertedComp against age
grouped_df.plot(kind='line', marker='o')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# Step 1: Count occurrences of each category in the 'MainBranch' column
main_branch_counts = df['MainBranch'].value_counts()

# Step 2: Create a horizontal bar chart
plt.figure(figsize=(10, 6))
main_branch_counts.plot(kind='barh', color='skyblue')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.title('Count of Main Branches')
plt.grid(axis='x')
plt.show()
```
